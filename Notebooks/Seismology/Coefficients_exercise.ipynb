# Reflection and transmission coefficients

#### SH Case

The reflection coefficient is

$
SS = \frac{\rho_1\beta_1 \cos(\theta_1) - \rho_2\beta_2 \cos(\theta_2)   }{ \rho_1\beta_1 \cos(\theta_1) + \rho_2\beta_2 \cos(\theta_2)     }
$

and the transmission coefficient

$
ST = \frac{2\rho_1\beta_1 \cos(\theta_1) }{ \rho_1\beta_1 \cos(\theta_1) + \rho_2\beta_2 \cos(\theta_2)     }
$

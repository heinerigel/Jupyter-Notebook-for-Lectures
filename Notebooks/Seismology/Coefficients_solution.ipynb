# Reflection and transmission coefficients

#### SH Case

The reflection coefficient is

$
SS = \frac{\rho_1\beta_1 \cos(\theta_1) - \rho_2\beta_2 \cos(\theta_2)   }{ \rho_1\beta_1 \cos(\theta_1) + \rho_2\beta_2 \cos(\theta_2)     }
$

and the transmission coefficient

$
ST = \frac{2\rho_1\beta_1 \cos(\theta_1) }{ \rho_1\beta_1 \cos(\theta_1) + \rho_2\beta_2 \cos(\theta_2)     }
$

**Exercise:** Plot the reflection and transmission behavior for $+/- 5.10,15,20,25,30,25,40\%$ shear velocity perturbation. Assume that $\beta_1$ = 3000m/s. Density is constant at 2500kg/$m^3$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# initialize incidence angles

n=100
eps = .01
th1 = np.linspace(eps,np.pi/2-eps,n)
ss = np.zeros(n)
st = np.zeros(n)

b1=3000.
r1=2500
r2=r1

for pert in [5,10,15,20,25,30,35,40]:

    b2 = (1-pert/100)*b1
    
    for i in range(n):

    
        th2=np.arcsin(np.sin(th1[i])*b2/b1)
    
        ss[i] = (r1*b1*np.cos(th1[i])-r2*b2*np.cos(th2))/(r1*b1*np.cos(th1[i])+r2*b2*np.cos(th2))

    plt.plot(th1/np.pi*180,np.absolute(ss))
plt.show()


for pert in [5,10,15,20,25,30,35,40]:

    b2 = (1+pert/100)*b1
    
    for i in range(n):

    
        th2=np.arcsin(np.sin(th1[i])*b2/b1)
    
        st[i] = (2*r1*b1*np.cos(th1[i]))/(r1*b1*np.cos(th1[i])+r2*b2*np.cos(th2))


    plt.plot(th1/np.pi*180,np.absolute(st))

plt.show()
```
